```python
# Count occurrences of each age group
age_group_counts = df_selected['Age groups'].value_counts()

# Get the count of 'Not reported' values
not_reported_count = age_group_counts.get('Not reported', 0)

# Calculate the total number of rows (without 'Not reported' rows)
total_count = len(df_selected) - not_reported_count

# Calculate the ratio for each age group (excluding 'Not reported')
age_group_ratios = age_group_counts.drop('Not reported') / total_count

# Generate a list of values to replace 'Not reported'
replacement_values = []

# Fill missing 'Not reported' values based on the calculated ratios
for _ in range(not_reported_count):
    selected_age_group = age_group_ratios.idxmax()  # Select the age group with the highest ratio
    replacement_values.append(selected_age_group)
    age_group_ratios[selected_age_group] -= 1  # Update the ratio to reflect the addition of a value

# Create a mask where the 'Age groups' column has 'Not reported' values
mask = df_selected['Age groups'] == 'Not reported'

# Replace 'Not reported' with the calculated replacements
df_selected.loc[mask, 'Age groups'] = replacement_values
```

```python
# Count occurrences of each age group
sex_group_counts = df_selected['Sex'].value_counts()

# Get the count of 'Not reported' values
not_reported_count = sex_group_counts.get('Not reported', 0)

# Calculate the total number of rows (without 'Not reported' rows)
total_count = len(df_selected) - not_reported_count

# Calculate the ratio for each age group (excluding 'Not reported')
sex_group_ratios = sex_group_counts.drop('Not reported') / total_count

# Generate a list of values to replace 'Not reported'
replacement_values = []

# Fill missing 'Not reported' values based on the calculated ratios
for _ in range(not_reported_count):
    selected_sex_group = sex_group_ratios.idxmax()  # Select the sex group with the highest ratio
    replacement_values.append(selected_sex_group)
    sex_group_ratios[selected_sex_group] -= 1  # Update the ratio to reflect the addition of a value

# Create a mask where the 'Age groups' column has 'Not reported' values
mask = df_selected['Sex'] == 'Not reported'

# Replace 'Not reported' with the calculated replacements
df_selected.loc[mask, 'Sex'] = replacement_values
```

```python
# Now you can apply the logic to df_test without affecting df_selected
# Run the redistribution logic on df_test
incorrect_rows = df_selected[(df_selected['Areas of expenditure'] == 'Dental expenditure') & (df_selected['Total Expenditure'] > 0)]

# Group by Year and sum the Total Expenditure for Dental expenditure (incorrect row)
total_expenditure_by_year = incorrect_rows.groupby('Year')['Total Expenditure'].sum().reset_index()
total_expenditure_by_year.rename(columns={'Total Expenditure': 'Cumulative Dental Expenditure'}, inplace=True)

# Merge the cumulative dental expenditure back into the original dataframe
df_selected = df_selected.merge(total_expenditure_by_year, on='Year', how='left')

# Calculate the proportion of each category within each year for the rows that are 'Dental expenditure'
category_counts = df_selected[df_selected['Areas of expenditure'] == 'Dental expenditure'].groupby(
    ['Year', 'Broad area', 'Age groups', 'Sex', 'Areas of expenditure']
).size().reset_index(name='Category Count')

# Merge this count into the dataframe to calculate proportions only for rows that are 'Dental expenditure'
df_selected = df_selected.merge(
    category_counts[['Year', 'Broad area', 'Age groups', 'Sex', 'Areas of expenditure', 'Category Count']],
    on=['Year', 'Broad area', 'Age groups', 'Sex', 'Areas of expenditure'], 
    how='left'
)

# Calculate the total count of categories for each year to get the proportion for each row
yearly_counts = df_selected.groupby('Year')['Category Count'].sum().reset_index(name='Total Count')
df_selected = df_selected.merge(yearly_counts, on='Year', how='left')

# Calculate the proportion of each category for each year
df_selected['Category Proportion'] = df_selected['Category Count'] / df_selected['Total Count']

# Redistribute the total expenditure across categories based on the proportions (only for 'Dental expenditure' rows)
df_selected['Redistributed Expenditure'] = df_selected['Category Proportion'] * df_selected['Cumulative Dental Expenditure']

# Update the 'Total Expenditure' column with the redistributed expenditure, but only for 'Dental expenditure' rows
df_selected.loc[df_selected['Areas of expenditure'] == 'Dental expenditure', 'Total Expenditure'] = df_selected['Redistributed Expenditure']

# Drop the temporary columns used for merging (cleanup)
df_selected.drop(['Cumulative Dental Expenditure', 'Category Count', 'Total Count', 'Category Proportion', 'Redistributed Expenditure'], axis=1, inplace=True)

# Now df_test has the redistributed values, and df_selected remains unchanged
```

```python
# Mapping function to match granular age groups from df_cleaned with broader categories in day_in_hospital
def map_age_group(age_group):
    if age_group in ['0-4', '5-9', '10-14']:
        return '0-14'
    elif age_group in ['15-19','20-24','25-29','30-34','35-39','40-44','45-49','50-54','55-59','60-64']:
        return '15-64'
    elif age_group in ['65-69','70-74','75-79','80-84','85+']:
        return '65+'
    return age_group

df_cleaned['Mapped Age Group'] = df_cleaned['Age Group'].apply(map_age_group)
```

```python
import plotly.graph_objects as go
import pandas as pd

# Aggregate the data by Age Group and Gender
aggregated_data = df_cleaned.groupby(['Age Group', 'Gender']).agg({'Expenditure per Person for a day': 'sum'}).reset_index()

# Create the figure
fig = go.Figure()

# Add males data as bars
males_data = aggregated_data[aggregated_data['Gender'] == 'Males']
fig.add_trace(go.Bar(
    x=males_data['Age Group'], 
    y=males_data['Expenditure per Person for a day'],
    name='Males',
    text=males_data['Expenditure per Person for a day'],
    textposition='auto',
    marker=dict(color='blue'),  # Set the color for males
    hovertemplate='%{x}: $%{y:,.2f}<extra></extra>',  # Format hovertext for males
    texttemplate='%{text:,.2f}'  # Format the text label on the bars
))

# Add females data as lines
females_data = aggregated_data[aggregated_data['Gender'] == 'Females']
fig.add_trace(go.Scatter(
    x=females_data['Age Group'],
    y=females_data['Expenditure per Person for a day'],
    mode='lines+markers',  # Adds both lines and markers
    name='Females',
    line=dict(width=2, color='red'),  # Set the line color for females
    marker=dict(size=8, color='red'),  # Set the marker color for females
    hovertemplate='%{x}: $%{y:,.2f}<extra></extra>'  # Format hovertext for females
))

# Customize the layout
fig.update_layout(
    title='Expenditure per Person for a Day by Age Group and Gender',
    xaxis_title='Age Group',
    yaxis_title='Expenditure per Person for a Day',
    barmode='group',  # Group the bars
    template='ggplot2'  # Dark theme for the plot
)

# Show the plot
fig.show()
```
